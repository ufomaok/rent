# Univariate Exploratory Analysis

Let's explore the distribution of single variables in this dataset.

```python
# Plot histograms, boxplots, and barcharts of all available numeric columns. 
# article: https://www.statology.org/univariate-analysis-in-python/ 




df = pd.read_excel("realestate.xlsx")

col = ['X2 house age', 'X4 number of convenience stores','X5 latitude','X6 longitude','Y house price of unit area']

#BOXPLOT
for column in col:
    plt.figure(figsize=(10, 7)) 
    sns.boxplot(data=df, y=column, color='pink')
    plt.title(f'Boxplot Chart of {column}')
    plt.ylabel(column)
   

#HISTOGRAM

#House Age
fig = plt.figure(figsize=(20,5))
ax = fig.gca()
df.hist(column='X2 house age', grid=False, edgecolor='black', ax = ax, bins = 400)
plt.xlabel('House Age')
plt.title('Histogram of House Age')
plt.ylabel('Frequency')



#NOTE - 'Distance to Nearest MRT' variable, unable to produce graph, this variable came in a 'string' format and has been turned into float in the 'clean data' file.


#Number of convenience stores
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
df.hist(column='X4 number of convenience stores', grid=False, edgecolor='black', ax = ax, bins = 150)
plt.xlabel('Number of Convenience Stores')
plt.title('Histogram of Number of Convenience Stores')
plt.ylabel('Frequency')

#Latitude
#NOTE - THE LATITUDE VARIABLE IS IRRELEVANT TO ANALYSIS, COLUMN/GRAPH HAS BEEN REMOVED IN THE 'CLEAN DATA' FILE. PLOTTING GRAHPH ONLY TO VISUALIZE DATA. 
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
df.hist(column='X5 latitude', grid=False, edgecolor='black', ax = ax, bins = 50)
plt.xlabel('Latitude')
plt.title('Histogram of Latitude')
plt.ylabel('Frequency')


#NOTE - THE LONGITUDE VARIABLE IS IRRELEVANT TO ANALYSIS, COLUMN/GRAPH HAS BEEN REMOVED IN THE 'CLEAN DATA' FILE. PLOTTING GRAHPH ONLY TO VISUALIZE DATA. 
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
df.hist(column='X6 longitude', grid=False, edgecolor='black', ax = ax, bins = 50)
plt.xlabel('Longitude')
plt.title('Histogram of Longitude')
plt.ylabel('Frequency')

#House Price of Unit Area
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
df.hist(column='Y house price of unit area', grid=False, edgecolor='black', ax = ax, bins = 50)
plt.xlabel('House Price of Unit Area')
plt.title('Histogram of House Price of Unit Area')
plt.ylabel('Frequency')
```

# Bivariate Exploratory Analysis

Let's explore the relationship of multiple variables in this dataset.

```python
# Pot scatter-plots and other visualizations of all available numeric columns. 
# article: https://www.statology.org/bivariate-analysis-in-python/



#Scatterplot of house age vs. price 
X2 = df['X2 house age']
Y = df['Y house price of unit area']

plt.figure(figsize=(10, 5))
plt.scatter(X2, Y)

plt.xlabel('House Age')
plt.ylabel('House Price of Unit Area')
plt.title(f'Scatterplot for House Age vs. House Price of Unit Area')

plt.grid(True)
plt.xlim(-10, 100)

#Scatterplot of distance to the nearest MRT station (CHANGE STRING TO INT)

X3 = df['X3 distance to the nearest MRT station']
Y = df['Y house price of unit area']

plt.figure(figsize=(20, 5))
plt.scatter(X3, Y)

plt.xlabel('Distance to nearest MRT Station')
plt.ylabel('House Price of Unit Area')
plt.title(f'Scatterplot for Distance to Nearest MRT Station vs. House Price of Unit Area')
plt.grid(True)
plt.xticks(rotation= 45)
plt.xlim(-.5, 50)

#Scatterplot of number of convenience stores -- X4 number of convenience stores 

X4 = df['X4 number of convenience stores']
Y = df['Y house price of unit area']

plt.figure(figsize=(20, 5))
plt.scatter(X4, Y)

plt.xlabel('Number of Convenience Stores ')
plt.ylabel('House Price of Unit Area')
plt.title(f'Scatterplot for Number of Convenience Stores vs. House Price of Unit Area')
plt.grid(True)
plt.xticks(rotation= 45)
plt.xlim(0, 110)

#Scatterplot latitude

X5 = df['X5 latitude']
Y = df['Y house price of unit area']

plt.figure(figsize=(20, 5))
plt.scatter(X5, Y)

plt.xlabel('Latitude')
plt.ylabel('House Price of Unit Area')
plt.title(f'Scatterplot for Latitude vs. House Price of Unit Area')
plt.grid(True)
plt.xticks(rotation= 45)
plt.xlim(0, 50)


#Scatterplot longitude

X6 = df['X6 longitude']
Y = df['Y house price of unit area']

plt.figure(figsize=(20, 5))
plt.scatter(X6, Y)

plt.xlabel('Longitude')
plt.ylabel('House Price of Unit Area')
plt.title(f'Scatterplot for Longitude vs. House Price of Unit Area')
plt.grid(True)
plt.xticks(rotation= 45)
plt.xlim(0, 200)
```
